** Visualization of data**
**bar-graph b/w sub-category and category**

```python
plt.figure(figsize=(16,8))
plt.bar("Sub-Category","Category",data=df)
plt.show()
```

**pie-plot of sub-category**

```python
plt.figure(figsize=(12,10))
df["Sub-Category"].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

**Bar plot of sales v/s profits**

```python
df.groupby("Sub-Category")["Profit","Sales"].agg(["sum"]).plot.bar()
plt.title("Total Profits and Sales as per Sub-Category")
#plt.Legend("Profit")
#plt.Legend("Sales")
plt.rcParams["figure.figsize"]=[10,10]
plt.show()
```

**count plot of cities**

```python
print(df["State"].value_counts())
plt.figure(figsize=(15,9))
sns.countplot(x=df["State"])
plt.xticks(rotation=90)
plt.show()
```

**Sales and Profit region wise**

```python
df.groupby("Region")["Profit","Sales"].agg(["sum"]).plot.bar()
plt.title("Total Profit and Sales per region")
#plt.Legend("Profit")
#plt.Legend("Sales")
plt.rcParams["figure.figsize"]=[8,8]
plt.show()
```

**count plot of sub-categories**

```python
print(df["Sub-Category"].value_counts())
plt.figure(figsize=(12,6))
sns.countplot(x=df["Sub-Category"])
plt.xticks(rotation=90)
plt.show()
```

**Heap map of corelation among columns**

```python
flg,axes=plt.subplots(1,1,figsize=(9,6))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

**Heat map of covariance among the set of variables**

```python
fig,axes=plt.subplots(1,1,figsize=(9,6))
sns.heatmap(df.cov(),annot=True)
plt.show()
```
